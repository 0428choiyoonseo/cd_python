- super( ).메서드( ) : 기반 클래스의 메서드 호출 <br>
    super(파생클래스, self).메서드 - 현재 클래스가 어떤 클래스인지 명확하게 표시

```python
class Person:
    def __init__(self):
        print('Person__init__')
        self.hello = '안녕하세요.'      #인스턴스 속성 생성해 문자열 저장
        
class Student(Person):      #Person class 상속받아 Student 클래스 생성
    def __init__(self):
        print('Student __init__')
        super().__init__()      #super()로 기반 클래스의 __init__ 메서드 호출. Person.__init__()과 같은 의미.
        self.school = '파이썬 코딩 도장'
        
james = Student()
print(james.school)     
print(james.hello)   #부모의 init을 호출해야 부모 속성(hello) 사용 가능          
```

    -> super() 사용 안해도 됨

# 메서드 오버라이딩 
: 파생 클래스에서 기반 클래스의 메서드를 새로 정의  <br>
    -원래 기능을 유지하면서 새 기능을 덧붙일 때 사용 <br>
    -기반 클래스의 메서드 무시하고 새 매서드 생성 

 -중복되는 기능은 파생클래스에서 재생성 x, 기반 클래스의 기능 사용

```python
class Person:
    def greeting(self):
        print('안녕하세요.')
        
class Student(Person):
    def greeting(self):
        super().greeting()      #기반 클래스의 greeting 메서드 호출하여 중복을 줄임
        print('안녕하세요. 저는 파이썬 코딩 도장 학생입니다.')
        
james = Student()
james.greeting()
```

# 추상 클래스(abstract class)

 : 메서드의 목록만 가진 클래스. 상속 받는 클래스에서 메서드 구현을 강제하기 위해 사용

    -추상 클래스의 추상 메서드를 모두 구현했는지 확인하는 시점은 파생 클래스가 인스턴스를 만들 때임 <br>
    -인스턴스로 만들 수 없음 (만들면 에러)



```python
from abc import *       #abc 모듈의 모든 클래스와 메서드 가져옴

class StudentBase(metaclass=ABCMeta):
    @abstractmethod
    def study(self):
        pass
 
    @abstractmethod
    def go_to_school(self):
        pass
 
class Student(StudentBase):
    def study(self):
        print('공부하기')
 
    def go_to_school(self):
        print('학교가기')
 
james = Student()    # 이 때 추상 클래의 추상 메서드를 모두 구현했는지 확인
james.study()
james.go_to_school()
```

```python
class AdvancedList(list):        # 리스트 상속       
    def replace(self, old, new):    #replace 메서드 이용. 클래스의 메서드 안에서 현재 객체 조작은 self 이용. AdvancedList가 list를 상속받으므로 self로 리스트의 모든 메서드 이용 가능
        while old in self:          # while로 반복하면서 self에 특정 요소가 있을 때 계속 반복하도록 만든 뒤 요소 변경
            self[self.index(old)] = new         # 리스트에서 가장 처음 나온 old의 위치 반환. index(list,index메서드)로 찾은 인덱스를 self에 지정해준 뒤 새 값 할당해 값 변경
                        #self.index(old) = list.index(self, old)
x = AdvancedList([1, 2, 3, 1, 2, 3, 1, 2, 3])
x.replace(1, 100)
print(x)
```

1. try except <br>
 try: 실행할 코드<br>

2. else 사용
 try: 실행할 코드<br>
 except: 예외가 발생했을 떄 처리하는 코드<br>
 else: 예외가 발생하지 않았을 때 실행할 코드 

 *else는 except 뒤에만 사용 가능

 3. finally
 try: 실행할 코드<br>
 except: 예외가 발생했을 떄 처리하는 코드<br>
 else: 예외가 발생하지 않았을 때 실행할 코드 <br>
 finally: 예외 발생 여부와 상관없이 항상 실행할 코드

 *finally는 except와 else 생략 가능

 

```python
try:
    x = int(input('나눌 숫자를 입력하세요: '))
    y = 10 / x
except ZeroDivisionError:    # 숫자를 0으로 나눠서 에러가 발생했을 때 실행됨
    print('숫자를 0으로 나눌 수 없습니다.')
else:                        # try의 코드에서 예외가 발생하지 않았을 때 실행됨
    print(y)
finally:                     # 예외 발생 여부와 상관없이 항상 실행됨
    print('코드 실행이 끝났습니다.')
```

- 예외 발생 시키기 (에러메시지 생략 가능) <br>

raise 예외('에러메시지')

```python
try:
    x = int(input('3의 배수를 입력하세요: '))
    if x % 3 != 0:
        raise Exception('3의 배수가 아닙니다.')
    print(x)
except Exception as e:
    print('예외 발생', e)    
```

현재 예외를 다시 발생시킴 -> except 안에서 raise 사용

- 예외 종류

- 예외 만들기 <br>
    -> Exception 상속 받아 새 클래스 생성. init 메서드 호풀하면서 에러 메시지 넣어주기

```python
class NotThreeMultipleError(Exception):    # Exception을 상속받아서 새로운 예외를 만듦
    def __init__(self):
        super().__init__('3의 배수가 아닙니다.')
 
def three_multiple():
    try:
        x = int(input('3의 배수를 입력하세요: '))
        if x % 3 != 0:                     # x가 3의 배수가 아니면
            raise NotThreeMultipleError    # NotThreeMultipleError 예외를 발생시킴
        print(x)
    except Exception as e:
        print('예외가 발생했습니다.', e)
 
three_multiple()
```
