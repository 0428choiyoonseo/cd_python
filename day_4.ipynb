# 25. 딕셔너리 응용
 딕셔너리명.setdefault(키, 기본값) : 딕셔너리 키-값 쌍 추가 후 해당 값 반환. 키만 지정하면 값에 None을 저장.  <br>
 딕셔너리명.update(키=값) : 키의 값 수정, 키가 없으면 키-값 쌍 추가. 괄호에 콤마로 구분하면 값 한꺼번에 수정 가능 <br>
 딕셔너리명.update(딕셔너리) 도 수정 가능

```python
 y = {1: 'one', 2: 'two'}
 y.update([[2, 'TWO'], [4, 'FOUR']]) #리스트 이용해 수정 가능
 y
 
```

```python
y.update(zip([1, 2], ['one', 'two'])) #update(반복가능한 객체)로 값 수정
y
```
