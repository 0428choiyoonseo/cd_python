## 18. break, continue -> 반복문 제어
	break : for / while 의 제어흐름 벗어나기 위함. 루프 중단
	continue :  제어흐름(반복) 유지한 상태에서 그 아래 코드 실행만 건너뜀

```python
for i in range(10):
    if i % 2 ==0:
        continue     #아래 코드를 실행하지 않고 건너뜀
    print(i)
```

pass :  아무 일도 안 함

```python
count = int(input('반복할 횟수를 입력하세요: '))
i=0
while True : 
    print(i)
    i+=1
    if i == count:
        break
```

## 19.계단식으로 별 출력
 바깥쪽 루프-세로, 안쪽 루프- 가로 방향 처리

```python
for i in range (3):      # 3번 반복, 세로 방향
    for j in range(5):   # 5번 반복, 가로 방향
        print('*', end='')  
    print()              # print는 출력 후 기본적으로 다음 줄로 넘어감
```

```python
for i in range(5):        
    for j in range(5):   
        if j <= i:                
            print('*', end='')    
    print()
    
    # 별 개수는 세로 방향에 비례하기에 i 에 맞추기
```

```python
for i in range(5):        
    for j in range(5):
        if j == i :
            print('*', end=' ')
        else:
            print(' ', end="")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if i<=j:
            print('*', end=" ")
        else:
            print(" ", end=" ")
    print()
```
